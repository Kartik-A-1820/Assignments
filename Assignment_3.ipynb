# 1.1 Write a Python Program to implement your own myreduce() function which works exactly  like Python's built-in function reduce() 

```python
#Creating reduce function
def myreduce(user_fun,user_iter):
    initial=user_iter[0]
    for i in user_iter[1:]:
        initial=user_fun(initial,i)
    return initial
```

```python
#Defining a adding function
def add(a,b):
    return a+b
#Define Max function
def returnMax(a,b):
    if(a>b):
        return a
    else:
        return b
#Define Max function
def returnMin(a,b):
    if(a<b):
        return a
    else:
        return b
#Define Average function using My_Reduce function
def returnAvg(list_var):
    return (myreduce(add,list_var))/len(list_var)
#list
l=[12,33,23,63,3,5,42,33,94,97]

#to find sum of all elements in the list
sumis=myreduce(add,l)
maxis=myreduce(returnMax,l)
minis=myreduce(returnMin,l)
```

```python
print('Sum of the list elements=',sumis)
print('Maximum element in the given list is',maxis)
print('Minimum element in the given list is',minis)
print('Average of the elements in the list:',returnAvg(l))
```

# 1.2 Write a Python program to implement your own myfilter() function which works exactly  like Python's built-in function filter() 

```python
#Defining user filter function
def myFilter(user_fun,user_iter):
    boolean=list()
    filtered_list=list()
    for i in user_iter:
        boolean.append(user_fun(i))
    for i in range(len(boolean)):
        if boolean[i]==True:
            filtered_list.append(user_iter[i])
    return filtered_list
            
```

```python
#defining function to Check even or odd
def checkEven(num):
    if(num%2 == 0):
        return True
    else:
        return False
def checkOdd(num):
    if(num%2 != 0):
        return True
    else:
        return False
#defining function to filter vowels from list of characters
def checkVowels(ch):
    ch=ch.lower()
    if ch in ['a','e','i','o','u']:
        return True
    else:
        return False
#Function to check positive and Negative numbers
def checkPositive(num):
    if num>0:
        return True
    else:
        return False
def checkNegative(num):
    if num<0:
        return True
    else:
        return False
```

```python
#list of characters
l=['a','b','c','d','e','f','a','o','e','w','s','r','u','s','z','u','k','a','r','t','i','k']
#list of numbers
nums=[1,2,3,56,-23,6,2,7,4,2,6,-2,36,25,355,656,-56,23,36,2,-5,3,47,-4,3,6,-4,-74,3,6,547,36,3,623,41,44,25,3,6,7]

print('The vowels in the given list=', myFilter(checkVowels,l))
print('The Even numbers in the given list=',myFilter(checkEven,nums))
print('The Odd numbers in the given list=',myFilter(checkOdd,nums))
print('The positive numbers in the given list of numbers=',myFilter(checkPositive,nums))
print('The negative numbers in the given list of numbers=',myFilter(checkNegative,nums))
```

# 2. Implement List comprehensions to produce the following lists.  
Write List comprehensions to produce the following Lists 
['A', 'C', 'A', 'D', 'G', 'I', ’L’, ‘ D’] 
['x', 'xx', 'xxx', 'xxxx', 'y', 'yy', 'yyy', 'yyyy', 'z', 'zz', 'zzz', 'zzzz']  
['x', 'y', 'z', 'xx', 'yy', 'zz', 'xxx', 'yyy', 'zzz', 'xxxx', 'yyyy', 'zzzz']  
[[2], [3], [4], [3], [4], [5], [4], [5], [6]] [[2, 3, 4, 5], [3, 4, 5, 6],  
[4, 5, 6, 7], [5, 6, 7, 8]] [(1, 1), (2, 1), (3, 1), (1, 2), (2, 2), (3, 2), (1, 3), (2, 3), (3, 3)]
 

```python
#1 ['A', 'C', 'A', 'D', 'G', 'I', ’L’, ‘ D’]
string = "ACADGILD"
list_ch = [ i for i in string ]
print ("ACADGILD => " + str(list_ch))
```

```python
#2 ['x', 'xx', 'xxx', 'xxxx', 'y', 'yy', 'yyy', 'yyyy', 'z', 'zz', 'zzz', 'zzzz']
string ='X Y Z'
list_ch=string.split()
result = [ i*n for i in list_ch for n in range(1,5)  ]
print(str(result))
```

```python
#3 ['x', 'y', 'z', 'xx', 'yy', 'zz', 'xxx', 'yyy', 'zzz', 'xxxx', 'yyyy', 'zzzz']
ch_list = ['x','y','z']
result = [ i*n for n in range(1,5) for i in ch_list  ]
print(str(result))
```

```python
#4 [[2], [3], [4], [3], [4], [5], [4], [5], [6]]
ch_list = [2,3,4]
result = [ [i+n] for i in ch_list for n in range(0,3)]
print(str(result))
```

```python
#5 [[2, 3, 4, 5], [3, 4, 5, 6],[4, 5, 6, 7],[5, 6, 7, 8]
ch_list = [2,3,4,5]
result = [ [i+n for i in ch_list] for n in range(0,4)  ]
print(str(result))
```

```python
#6 [(1, 1), (2, 1), (3, 1), (1, 2), (2, 2), (3, 2), (1, 3), (2, 3), (3, 3)]
input_list=[1,2,3]
result = [ (b,a) for a in input_list for b in input_list]
print(str(result))
```
